```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=[1,5,6,7,3,2]
y=[1,2,3,4,5,6]

plt.xlim(0, 10)  # Set x-axis limits from 2 to 8
plt.ylim(0, 7)
plt.plot(x,label='x')
plt.plot(y,label='y')
plt.autoscale(axis='both')

plt.legend()
plt.show()
```

```python
x1=[1,2,3,4,5]
y1=[6,5,4,4,3]
x2=[8,7,6,5,4]
y2=[3,4,5,6,7]
plt.figure(figsize = (5, 6), dpi = 100)
plt.xlim(0, 10)  # Set x-axis limits from 2 to 8
plt.ylim(0, 7)

plt.plot(x1, y1, color = 'purple', lw = 2.5)
plt.plot(x2, y2, color = 'green', lw = 2.5)

plt.show()
```

```python
plt.plot([1,2,3,4,5], [1,4,9,16, 25], color = 'red', lw = 2.5, marker = '*', mec = 'green', mew = 5)

plt.show()
```

```python
x = np.array([1,2,3,4,5])

y1 = x**2
y2 = np.exp(x)
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r')

plt.subplot(1,2, 2)
plt.plot(x, y2, 'b-.')
```

```python
x=[9,8,7,6,5,4]
y=[1,2,3,4,5,6]
z=[5,6,4,7,3,8]
plt.figure(figsize=(10,5) )
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,z,'b')
plt.show()
```

```python
fig = plt.figure(figsize = (10,6))
axes1 = fig.add_axes([0.11, 0.22, 0.8, 0.8]) # Corrected arguments
axes2 = fig.add_axes([0.20, 0.34, 0.3, 0.3]) # Corrected arguments

axes1.plot(x,y, 'r--')
axes1.set_title('AAAYYYYOOOO')

axes2.plot(y,x, 'g-.')
axes2.set_title('AYYO')
```

```python
fig=plt.figure(figsize=(10,5))
axes=fig.subplots(2,3)
axes=axes.flatten()
for i in range(6):
  axes[i].plot(x,y)
plt.show()
```
