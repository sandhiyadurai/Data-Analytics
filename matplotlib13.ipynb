```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([5,6,7,8,5])
y=np.array([10,2,3,4,5])
z=np.array(x*y)
plt.title('Ayyyyyyyyooooooo')
plt.plot(x,color='green' ,linewidth=5,linestyle='dotted')
plt.plot(y,color='black',linewidth=6,linestyle='dashed')
plt.plot(z,color='red',linewidth=5)
plt.show()
```
