```python
import pandas as pd
import numpy as np

data={
    'name':["gorila","elephent","cat","dog"],
    'class' :[4,5,5,3],
    'weight':[100,200,10,12]
}
```

```python
data
```

```python
df=pd.DataFrame(data)#----- dATAFRAME----
df
```

```python
#--- GROP BY---#
MY=df.groupby('name')
```

```python
MY.mean()  #---- mean-----
```

```python
MY.std() #---- std-----
```

```python
MY.median() #---- median---
```

```python
df.groupby('weight').describe() #-------descrobe----
```

```python
MY.sum()
```
